```python
# Panda
# ------------------------------------------------------------
# Avec  Numpy  et  Matplotlib, la librairie  Pandas  fait partie des librairies de base pour la data science 
# Pandas fournit des structures de données puissantes et simples à utiliser, 
# ainsi que les moyens d'opérer rapidement des opérations sur ces structures. 
```

```python
# Imaginons que nous souhaitions représenter des pandas !
# Nous pouvons caractériser un panda par ses tailles de pattes, de longueur moyenne des poils de sa fourrure, 
# de queue, ainsi que le diamètre de son ventre.
```

```python
# Représentation par un tableau numpy
# ------------------------------------------------------------

import numpy as np
un_panda_numpy = np.array([100,5,20,80])
un_panda_numpy

# Ici, le panda a des pattes de 100cm, des poils de 5cm en moyenne, 
# une queue de 20cm et un ventre de 80cm de diamètre.
```

```python
# Pour représenter plusieurs pandas, je crée un tableau multi-dimensionnel  

famille_panda = [
    [100, 5  , 20, 80], # maman panda
    [50 , 2.5, 10, 40], # bébé panda
    [110, 6  , 22, 80], # papa panda
]
famille_panda_numpy = np.array(famille_panda)
famille_panda_numpy

# Ici, nous avons créé une liste de pandas, c'est à dire une liste de listes.
# Le rang de famille_panda_numpy est de 2, car nous avons 2 niveaux d'imbrication.
```

```python
# QUESTTION : Trouver l'instruction pour obtenir la taille des pattes 
# (situé en position 0 dans chaque liste qui décrit un panda) du 2e panda

# REPONSE : 
famille_panda_numpy[1, 0] 

```

```python
# QUESTTION : Trouver l'instruction pour obtenir les tailles des pattes de toute la famille panda ?

# REPONSE : 
famille_panda_numpy[:, 0] 
```

```python
# Représentation avec la librairie Pandas
# ------------------------------------------------------------
# Tout est une histoire de tableau !

famille_panda = [
    [100, 5  , 20, 80],
    [50 , 2.5, 10, 40],
    [110, 6  , 22, 80],
]
# ... ça ressemble un peu à un tableau avec des lignes et des colonnes ! 
```

```python
import pandas as pd
famille_panda_df = pd.DataFrame(famille_panda)
famille_panda_df
```

```python
# L'objet qui permet de représenter des "tableaux" est l'objet DataFrame. 
# Pour instancier un tel objet, on lui transmet une liste de rang 2, 
# c'est à dire une liste de listes.

# Nous pouvons même faire mieux, en indiquant les noms de colonnes et des lignes. 
# Et encore mieux : la librairie Pandas se base en grande partie sur la librairie Numpy 
# dans son fonctionnement interne. On peut très bien transmettre 
# à l'objet DataFrame de la donnée au format ndarray :
```

```python
famille_panda_df = pd.DataFrame(famille_panda_numpy,
                                index = ['maman', 'bebe', 'papa'],
                                columns = ['pattes', 'poil', 'queue', 'ventre'])
famille_panda_df
```

```python
# Le nom des lignes est appelé index. Un index peut être une chaîne de caractères (un label) ou un nombre entier. 
# Quand aucun index n'est spécifié à la création du dataframe, il est initialisé par défaut avec une suite continue 
# d'entiers commençant par 0.
```

```python
# Quel est l'intérêt de cette librairie  ?
# L'objet DataFrame est similaire :
# - aux tables des bases de données relationnelles que l'on manipule grâce au langage SQL
# - à l'objet dataframe sur lequel se base tout le langage R
# Du coup, si vous connaissez déjà les langages SQL ou R, vous aurez beaucoup de facilité à utiliser 
# le DataFrame de Pandas !
```

```python
# Quelques fonctionnalités des DataFrames
# Tout d'abord, accédons à la colonne ventre de notre table. Il y a deux syntaxes possibles, 
# qui renvoient exactement le même résultat :

famille_panda_df.ventre

```

```python

famille_panda_df["ventre"]
```

```python
# QUESTION : Afficher lalongeur des poils des pandas

# REPONSE : 
famille_panda_df.poil
```

```python
# L'objet que renvoie famille_panda_df["ventre"] est de type pandas.Series. 
# Pour obtenir les valeurs de la colonne ventre au format numpy, 
# il faut saisir famille_panda_df["ventre"].values

# Parcourrons maintenant tous les pandas un à un, grâce à la méthode iterrows qui renvoie 
# (à chaque itération de la boucle for) un tuple dont le premier élément est l'index de la ligne, 
# et le second le contenu de la ligne en question :

for ind_ligne, contenu_ligne in famille_panda_df.iterrows():
    print("Voici le panda %s :" % ind_ligne)
    print(contenu_ligne)
    print("--------------------")
```

```python
# Accédons maintenant au papa panda : d'abord par sa position (2), puis par son nom "papa". 
# Le résultat retourné est exactement le même dans les 2 cas.
```

```python
famille_panda_df.iloc[2] # Avec iloc(), indexation positionnelle
```

```python
famille_panda_df.loc["papa"] # Avec loc(), indexation par label
```

```python
# Déterminons les pandas dont le diamètre du ventre est de 80cm :
famille_panda_df["ventre"] == 80
# Ici, on teste chaque élément de la colonne "ventre" en demandant s'il est égal à 80. 
# On obtient la réponse suivante : "maman" : True, "bebe" : False, "papa" : True.
```

```python
# Le résultat de cette opération est très pratique pour filtrer des lignes ! 
# Par exemple, pour sélectionner uniquement les pandas dont le ventre est de 80cm, 
# il suffit d'intégrer ce précédent résultat en tant que masque, comme ceci :

masque = famille_panda_df["ventre"] == 80
pandas_80 = famille_panda_df[masque]

pandas_80
```

```python
# Qu'est-ce qu'un masque ?
# Lorsque vous mettez un masque sur votre visage, celui-ci cache certaines parties de votre tête, 
# mais en laisse visible d'autres : vos yeux, votre nez et votre bouche. 
# Ici, nous souhaitons  garder uniquement certaines lignes du dataframe, et en cacher d'autres. 
# Un masque est ici une liste de variables booléennes (True ou False) 
# dans laquelle chaque élément est associé à une ligne du dataframe. 
# Si cet élément est True, cela signfie que nous souhaitons garder la ligne sinon cet élément est False.

# QUESTION : afficher les informations sur les pandas dont la longueur des poils est supérieure à 4

# REPONSE :
masquePoils = famille_panda_df["poil"] > 4
pandas_sup4 = famille_panda_df[masque]

pandas_sup4
```

```python
# Pour inverser le masque, il suffit d'utiliser l'opérateur ~, et nous sélectionnons les pandas qui n'ont pas un ventre de 80cm :
famille_panda_df[~masque]
```

```python
# QUESTION : afficher les informations sur les pandas dont la longueur des poils est inpérieure à 4 avec l'opérateur ~

# REPONSE :
famille_panda_df[~masquePoils]
```

```python
# Maintenant, ajoutons des lignes à notre dataframe. 
# Il y a plusieurs méthodes pour cela, mais voyons ici la plus simple : assemblons ensemble deux dataframes.
```

```python
quelques_pandas = pd.DataFrame([[105,4,19,80],[100,5,20,80]],      # deux nouveaux pandas
                               columns = famille_panda_df.columns) 
                               # même colonnes que famille_panda_df
tous_les_pandas = famille_panda_df.append(quelques_pandas)
tous_les_pandas
```

```python
# Dans le dataframe tous_les_pandas, il y a des doublons. 
# Le premier panda (maman) et le dernier  (dont l'index est 1) ont exactement les mêmes mesures. 
# Si nous souhaitions dédoublonner, nous ferions ceci :
```

```python
tous_les_pandas.drop_duplicates()
```

```python
# Le résultat de cette ligne ne modifie pas le dataframe  tous_les_pandas  . 
# Il ne fait que renvoyer un autre dataframe. 
# Dans l'état actuel, ce nouveau dataframe est juste affiché et n'est pas enregistré en mémoire : 
# il sera donc perdu. 
# Pour le garder, on peut
# - soit l'enregistrer dans une nouvelle variable : pandas_uniques = tous_les_pandas.drop_duplicates()
# - soit remplacer la variable tous_les_pandas par ce nouveau dataframe dédoublonné :
#               tous_les_pandas = tous_les_pandas.drop_duplicates()
```

```python
# Données manquantes
# --------------------------------------------------------------------
# Vous aurez remarqué, dans la sortie de la fonction  describe la présence de valeurs  NaN. 
# C'est une valeur définie pour représenter quelque chose qui n'est pas un nombre (Not a Number) 
# alors que son type l'exige. 
# Plus généralement, le résultat de toute opération impliquant une NaN est à son tour un NaN. 

# Nous allons  voir deux opérations à appliquer aux valeurs manquantes NaN.

titanic.age.head(10)
```

```python
# QUESTION : Quel effet, la fonction ci-dessous a t'elle sur  les valeurs manquantes.

titanic.fillna(value={"age": 0}).age.head(10)

# REPONSE : Elle remplace la valeur NaN par 0 
```

```python
# QUESTION : Quel effet, la fonction ci-dessous a t'elle sur  les valeurs manquantes.
titanic.fillna(method="pad").age.head(10)
# REPONSE : 
Elle donne à la ligne NaN la valeur de la ligne du dessus
```

```python
# QUESTION : Quel effet, la fonction ci-dessous a t'elle sur  les valeurs manquantes.
titanic.dropna().head(10)
# REPONSE : elle affiche les 10 premières lignes qui n'ont aucune valeur NaN pour aucune colonne
```

```python
# QUESTION : Quel effet, la fonction ci-dessous a t'elle sur  les valeurs manquantes.
titanic.dropna(axis="columns").head()
# REPONSE : Elle enlève les colonnes où on aurait eu une valeur NaN
```

```python
# QUESTION : A quoi sert la fonction ci-dessous ?

titanic.rename(columns={"sex":"sexe"})
# REPONSE : Elle permet de renommer une colonne
```

```python
# La fonction  drop  permet de supprimer des axes (colonnes ou lignes) d'un DataFrame. 
titanic.drop(0)
# Supprimera la ligne dont l'index est égal à 0.

# QUESTION : Supprimer la colonne "age"
# REPONSE : titanic.drop(3)

# IMPORTANT: De nombreuses fonctions Pandas, acceptent l'argument  "inplace".  
# Si la valeur de cet argument est  True, le DataFrame donné en argument est modifié. 
# Sinon, une copie du DataFrame est retournée par la fonction.
```

```python
# Tableaux croisés dynamiques
# ----------------------------------------------------------
# Ces tableaux sont utilisés dans des logiciels tableurs.
# Ces tableaux permettent de synthétiser les données contenues dans un DataFrame. 

# QUESTION : A quoi sert la fonction pivot_table ?
titanic.pivot_table('survived', index='sex', columns='class')

# REPONSE :  Elle permet de créer un nouveau dataFrame concernant le premier paramètre, en choisissant d'autres 
# colonnes comme attributs
```

```python
# QUESTION : Quelle est le nombre total de survivants selon le sexe et le type du billet ? 
# Modifier la fonction précédente selon l'argument aggfunc="sum"
titanic.pivot_table(aggfunc="sum", index='sex', columns='class')

# REPONSE : 
```

```python
# QUESTION : comment interprétez vous le code ci-dessous ?
titanic.dropna(inplace=True)
age = pd.cut(titanic['age'], [0, 18, 80])
titanic.pivot_table('survived', ['sex', age], 'class')
# REPONSE : On commence par remplacer le dataframe par ce même dataframe mais sans les lignes ayant des valeurs à 
#NaN, puis on range les personnes selon leur âge dans deux intervalles puis on affiche le sexe des personnes ayant
# survécu en fonction du type de le billet, classés selon leur âge dans l'intervalle définit précédemment.
```
